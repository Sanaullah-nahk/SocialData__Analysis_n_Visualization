> * Explain in your own words: What is the problem with pie-charts?

<font face="Comic sans MS" size="3" color="blue">Pie charts are not a suitable visualization method when there are a large number of variables or the differences between the variables are not significant. If there are too many variables, the pie chart may have too many slices, making it difficult for the human eye to perceive clearly. It is recommended to use a pie chart only for three or fewer variables. Additionally, when the differences between the variables are not significant, it may be difficult to distinguish which variable has a larger slice on the pie chart compared to using a bar chart.

> We can call this kind of visualization a *point scatter plot*.

```python
df1 = df[ (df['Category']=='DRUG/NARCOTIC') &  (df['Date'].between('2016-06-01','2016-07-31' )) ]

poionts_map = folium.Map(location=[37.773972, -122.431297], tiles = "Stamen Toner", zoom_start=12)

for index, row in df1.iterrows():
    point = [row['Y'], row['X']]
    #folium.Marker(point).add_to(poionts_map)
    folium.CircleMarker(location=point, radius=1, color='red', fill_color='red').add_to(poionts_map)
    #folium.Marker(point, icon=folium.Icon(icon='Circle',color='blue', prefix='glyphicon',icon_size=(10,10))).add_to(poionts_map)
poionts_map
```

```python
from folium import plugins
from folium.plugins import HeatMap

df2 = df[ (df['Category']=='PROSTITUTION')]

poionts_map = folium.Map(location=[37.773972, -122.431297], tiles = "Stamen Toner", zoom_start=13)

heat_df = df2[['Y', 'X']].copy()

# Create weight column, using date
heat_df['Weight'] = df2['Date'].astype('str').str[5:7].astype(float)

# List comprehension to make out list of lists
heat_data = [[[row['Y'],row['X']] for index, row in heat_df[heat_df['Weight'] == i].iterrows()] for i in range(1,13)]

# Plot it on the map
hm = plugins.HeatMapWithTime(heat_data, auto_play=True, max_opacity=0.8, )

hm.add_to(poionts_map)
# Display the map
poionts_map
```

```python
df2 = df[ (df['Category']=='PROSTITUTION')]

poionts_map = folium.Map(location=[37.773972, -122.431297], tiles = "Stamen Toner", zoom_start=13)

heat_df = df2[['Y', 'X']].copy()

# Create weight column, using date
heat_df['Weight'] = df2['Date'].astype('str').str[5:7].astype(float)

# List comprehension to make out list of lists
heat_data = [[[row['Y'],row['X']] for index, row in heat_df[heat_df['Weight'] == i].iterrows()] for i in range(1,13)]

# Plot it on the map
hm = plugins.HeatMapWithTime(heat_data, auto_play=True, max_opacity=0.8, radius=9, blur=1)

hm.add_to(poionts_map)
# Display the map
poionts_map
```
