Explain in your own words how the Pearson correlation works and write down it's mathematical formulation. Can you think of an example where it fails (and visualization works)?
##### ANSWER:
<font face="Comic sans MS" size="2" color="blue">  The Pearson correlation measures the linear relationship between two continuous variables, with values ranging from -1 (perfect negative correlation) to 1 (perfect positive correlation), and 0 indicating no correlation.The mathematical formulation for Pearson correlation between two variables X and Y  can be represented as: 

$$
r=\frac{\sum\left(x_i-\bar{x}\right)\left(y_i-\bar{y}\right)}{\sqrt{\sum\left(x_i-\bar{x}\right)^2 \sum\left(y_i-\bar{y}\right)^2}}
$$

<font face="Comic sans MS" size="2" color="blue">  Pearson correlation fails is when there is a non-linear relationship between the variables. In such cases, the correlation coefficient may be close to zero even though there is a strong relationship between the variables. In such cases, visualization can help in identifying the non-linear relationship.
    
<font face="Comic sans MS" size="2" color="blue"> Let's say that x is the temperature of a room, and y is the number of people in the room. As the temperature increases, people might be more likely to leave the room, leading to a non-linear relationship.

<font face="Comic sans MS" size="2" color="blue"> For instance, if we plot the data and find that as the temperature increases beyond a certain point, the number of people in the room drops off sharply, we can conclude that there is a strong, non-linear relationship between temperature and the number of people in the room, even if the Pearson correlation coefficient suggests a weak or no correlation.

> * What is a *Quantile plot*? What is it good for. 

<font face="Comic sans MS" size="2" color="blue"> A quantile plot is a graphical way to display the distribution of a dataset.  It is constructed by sorting the data and then plotting each data point against its corresponding quantile. 
The basic idea behind a quantile plot is to compare the empirical quantiles of the data to the quantiles of the theoretical distribution. If the data follows the theoretical distribution, the points on the plot will form a straight line. Deviations from a straight line can indicate departures from the theoretical distribution.

<font face="Comic sans MS" size="2" color="blue">Quantile plots are useful for identifying departures from normality and for assessing the fit of a statistical model to the data. They can also be used to compare the distributions of two or more datasets.



> * I want to tell you about another interesting way to use box plots. To begin, we'll find out what time of day crimes usually happen for each type of crime. Then, we'll create a chart showing the average time of day that each type of crime happens. There are many ways to figure this out, but I'll explain one of them.". 
>   * "You only need to keep track of the time of day for each data point, so you can ignore everything else."
>   * Another challenge you need to address is that time doesn't divide neatly into units of 100 like many other numbers. I can think of two ways to deal with this.
>     * To represent each time of day, just count the number of seconds that have passed since midnight.
>     * Another way to represent each time of day is to round it to the nearest hour and then convert the minutes and seconds into a percentage of an hour. So 10:15 $\rightarrow$ 10.25, 8:40 $\rightarrow$ 8.67, etc.
> * You can use box plots to see when different crimes tend to happen. It's important to note that these plots have a different meaning than the ones we made in the previous exercise. Pretty cool, huh?  

```python
# Another way to represent each time of day is to round it to the nearest hour and then convert the minutes and seconds into a percentage of an hour. So 10:15  â†’
#  10.25, 8:40  â†’ 8.67, etc.

# You only need to keep track of the time of day for each data point, so ignore everything else.
temp = df[df['Category'].isin(focuscrimes)][['Category', 'Time']].sort_values('Category')
```

```python
# To represent each time of day, just count the number of seconds that have passed since midnight.
temp['Time'] = temp['Time'].dt.hour * 3600 +temp['Time'].dt.minute * 60 + temp['Time'].dt.second
```

```python
sns.set(style="white")

fig = plt.figure(figsize=(8,4), dpi=100)
sns.boxplot(x='Category', y='Time', data=temp)

plt.xticks(rotation=90, fontsize=10);
plt.ylabel('Time of Day (Seconds since Midnight)', fontsize=14);
plt.title('Time of Day for Different Crime Categories', fontsize=14);
plt.yticks(range(0, max(temp.Time)+10000, 10000), fontsize=10);
```
