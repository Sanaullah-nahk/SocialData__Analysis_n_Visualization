```python
data_link = 'D:\DTU\SP 2023\Social Data Viz\Data'
data = pd.read_csv(data_link+'\Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv')

df = data.iloc[:, :14] # frist 14 columns
```

```python
df.dtypes
```

```python
df.loc[:, 'Date'] = pd.to_datetime(df['Date']) # convert date column to datetime object
```

```python
# Filter date # 'YYYY-MM-DD'

start_date = pd.to_datetime('2003-01-01')
end_date = pd.to_datetime('2017-12-31')

df = df[ (df['Date'] >= start_date) & (df['Date'] <= end_date)]
```

> * Now generate the following simple statistics
>   - Report the total number of crimes in the dataset.
>   - List the various categories of crime. How many are there? 
>   - List the number of crimes in each category.

```python
#List the various categories of crime. How many are there?
unique_catagory = df["Category"].unique()
print(f'Total number of crime categories : {len(unique_catagory)}\n')
print(unique_catagory)
```

```python
# List the number of crimes in each category.
cat_count = df['Category'].value_counts()
print(cat_count)
```

In order to do awesome *predictive policing* later on in the class, we're going to dissect the SF crime-data quite thoroughly to figure out what has been going on over the last  years on the San Francisco crime scene. 

> ## Exercise 2
> The types of crimes. The first field we'll dig into is the column "Category".
> * We have already counted the number of crimes in each category. What is the most commonly occurring category of crime? What is the least frequently occurring?
> * Create a bar-plot over crime occurrences. This is a data visualization class, so here is the first essential lesson: **For a plot to be informative you need to label the axes** (The police chief will be furious if you forget). It can also be nice to other relevant pieces of info, title, labels, etc.)

* We have already counted the number of crimes in each category. What is the most commonly occurring category of crime? What is the least frequently occurring?

```python
most_common  = cat_count.idxmax()
least_frequent = cat_count.idxmin()

print(f'most commonly occurring category of crime:\t {most_common}')
print(f'east frequently occurring crime:\t\t {least_frequent}')
```

* Create a bar-plot over crime occurrences. 

```python
#Using Seaborn
fig = plt.figure(figsize=(10,5), dpi=100);

ax = sns.countplot(data=df, x='Category', order=df['Category'].value_counts().sort_values(ascending=False).index)

for i in ax.containers:
    plt.bar_label(i, rotation = 90, padding=5,color='navy')
plt.margins(x=0.0, y=0.2)
plt.title('Crime occurrence by catagory', fontsize=20)
plt.xticks(rotation = 90);
plt.xlabel('Catagories', fontsize=16);
plt.ylabel('Occurrences', fontsize=16);
```

```python
# Using Pandas visualization 
fig = plt.figure(figsize=(10,5), dpi=100);
ax = df['Category'].value_counts().plot(kind='bar')

for i in ax.containers:
    plt.bar_label(i, rotation = 90,  color='navy',padding=5)

plt.margins(x=0.0, y=0.2)


plt.title('Crime occurrence by catagory', fontsize=18)
plt.xticks(rotation = 90);
plt.xlabel('Category', fontsize=12);
plt.ylabel('Occurrences', fontsize=12);
```

The plot above already gives us a bit of information but does not tell us anything about how crimes changed over the years. Let's have a look at it!

> * To start off easily, let's count the number of crimes per year:

```python
df['year'] = df.Date.dt.year 
```

```python
df.year.value_counts()
```

```python
# OR same can be done by group by 
df.groupby('year').size().reset_index(name='counts')
```

>   *  What is the year with most crimes? 
>   * What is the year with the fewest crimes?

```python
max_year = df.year.value_counts().idxmax()
min_year = df.year.value_counts().idxmin()

print(f'year with most crimes:\t\t {max_year}')
print(f'year with fewest crimes:\t {min_year}')
```

> * Create a barplot of crimes-per-year (years on the $x$-axis, crime-counts on the $y$-axis).

```python
df.head()
```

```python
# Using Pandas visualization 
fig = plt.figure(figsize=(10,5), dpi=100);
ax = df['year'].value_counts().sort_index().plot(kind='bar')

for i in ax.containers:
    plt.bar_label(i, rotation = 90,  color='navy',padding=5,label_type='center')

plt.margins(x=0.0, y=0.1)


plt.title('Crime occurrence by year', fontsize=18)
plt.xticks(rotation = 90);
plt.xlabel('year', fontsize=12);
plt.ylabel('crime counts', fontsize=12);
```

```python
focuscrimes = set(['WEAPON LAWS', 'PROSTITUTION', 'DRIVING UNDER THE INFLUENCE', 'ROBBERY', 'BURGLARY', 'ASSAULT',
                   'DRUNKENNESS', 'DRUG/NARCOTIC', 'TRESPASS', 'LARCENY/THEFT', 'VANDALISM', 'VEHICLE THEFT', 
                   'STOLEN PROPERTY', 'DISORDERLY CONDUCT'])
```

```python
sns.countplot(data = df[df['Category']=='ASSAULT'], x='year')
```

```python
focuscrimes=sorted(focuscrimes, key=str.upper)
fig = plt.figure(figsize=(18,27), dpi=100);
plt.tight_layout()
fig.subplots_adjust(hspace=.3, wspace=.15)
for i,l in enumerate(focuscrimes):
    data1 = df[df['Category']==l]
    
    plt.subplot(7,2,i+1)
    ax = sns.countplot(data=data1, x='year', label = l)
    plt.margins(x=0.0, y=0.2)
    plt.legend(loc=2)
    plt.ylabel('Occurrences', fontsize=12)
```

```python
focuscrimes=sorted(focuscrimes, key=str.upper)
fig = plt.figure(figsize=(18,27), dpi=200);
plt.tight_layout()
fig.subplots_adjust(hspace=.4, wspace=.15)
for i,l in enumerate(focuscrimes):
    
    plt.subplot(7,2,i+1)
    df[df['Category']==l].year.value_counts().sort_index().plot(kind='bar')   
    #ax = sns.countplot(data=data1, x='year', label = l)
    plt.margins(x=0.0, y=0.3)
    plt.text(0,max(df[df['Category']==l].year.value_counts().sort_index())+1000, l)
    #plt.legend(loc=2)
    plt.ylabel('Occurrences', fontsize=10)
```
