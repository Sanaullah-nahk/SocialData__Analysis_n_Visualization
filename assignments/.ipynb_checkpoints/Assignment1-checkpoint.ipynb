```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv')
df_focus = df[df['Category'].isin(focuscrimes)].copy()
df_focus.head()
```

```python
df_focus_cat = df_focus.groupby('Category')
fig, axes = plt.subplots(7, 2, figsize=(20, 80))

index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
for (cat, group), ax in zip(df_focus_cat, axes.flatten()):
    group['DayOfWeek'].value_counts().reindex(index).plot(kind='bar', title=cat, ax=ax, xlabel='Day', ylabel='Occurences')

fig.suptitle('Bar plots of focus crimes weekday occurences', y=0.9, fontsize=16)

plt.show()
```

After observing the weekday plots, some crimes show an even spread of crime occurences over the week such as 'ROBBERY', 'STOLEN PROPERTY' and 'WEAPON LAWS'. Other crimes have a larger variation of occurences over the week. For example, 'DRUNKENESS' has about 500 and more occurences over the weekend as drinking activities are higher on the weekend. 'BURGLARY' has fewest crimes on the weekend, which might be due to more people being at home over the weekend. However, there are crimes with interesting variation of occurences such as 'PROSTITUTION' and 'DRUG/NARCOTIC'. The number of occurences for 'PROSTITUTION' on Thursday is significantly higher than the other days at about 3600 occurences. The number of occurences for 'DRUG/NARCOTIC' spikes up on Wednesday at 20000.

```python
df_focus['Month'] = df_focus['Date'].apply(lambda x: x[:2])
df_focus_cat = df_focus.groupby('Category')
fig, axes = plt.subplots(7, 2, figsize=(20, 80))

for (cat, group), ax in zip(df_focus_cat, axes.flatten()):
    group['Month'].value_counts().sort_index().plot(kind='bar', title=cat, ax=ax, xlabel='Month', ylabel='Occurences')

fig.suptitle('Bar plots of focus crimes monthly occurences', y=0.9, fontsize=16)

plt.show()
```

After observing the monthly occurences, we observed that most crimes show a variation in occurences over the months and are not evenly spread. We have found out that March has one of the highest number of crime occurences for most crimes. An interesting observation is that December has the lowest occurences of 'DISORDERLY CONDUCT' at about 700. We would have thought that December would have higher occurences as it is a festive month which brings more opportunities for 'DISORDERLY CONDUCT'

```python
df_focus['Hour'] = df_focus['Time'].apply(lambda x: x[:2])
df_focus_cat = df_focus.groupby('Category')
fig, axes = plt.subplots(7, 2, figsize=(20, 80))

for (cat, group), ax in zip(df_focus_cat, axes.flatten()):
    group['Hour'].value_counts().sort_index().plot(kind='bar', title=cat, ax=ax, xlabel='Hour', ylabel='Occurences')

fig.suptitle('Bar plots of focus crimes 24-hour cycle occurences', y=0.9, fontsize=16)

plt.show()
```

After observing the 24-hour cycle plot, we observed that most crimes have varying distributions where occurences are low across a certain time period followed by a signifcant increase for the next time period. However, a similarity that we observed is most crimes occur later in the day at about 1700 hours onwards. For example, 'VEHICLE THEFT' and 'VANDALISM' occurences start increasing significantly by about 2000 from 1700 hours onwards.

```python
df_focus['HourOfWeek'] = df_focus['Hour'] + ' - ' + df_focus['DayOfWeek']
df_focus_cat = df_focus.groupby('Category')
fig, axes = plt.subplots(7, 2, figsize=(20, 80), sharex=True)

index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
index_how = []
for day in index:
    for i in range(24):
        if i < 10:
            index_how.append('0' + str(i) + ' - ' + day)
        else:
            index_how.append(str(i) + ' - ' + day)

# Should be histogram
for (cat, group), ax in zip(df_focus_cat, axes.flatten()):
    group['HourOfWeek'].value_counts().reindex(index_how).plot(kind='bar', title=cat, ax=ax, xlabel='Hour - Day', ylabel='Occurences')

fig.suptitle('Bar plots of focus crimes 168-hour cycle occurences', y=0.9, fontsize=16)

plt.show()
```

After observing the 168-Hour cycle plot, we observed that most crimes have a similar distribution of occurences for each day of the week. This means that most crime occurences have a similar 24-hour pattern throughout the week.

What is the problem with pie-charts?

Pie-charts are not effective visulization techniques when there are many variables or the differences between variables are insignificant. When there are many variables, the pie chart will have too many slices for the human eye to observe clearly. Pie charts should be used for 3 or less variables. Secondly, when the differences between variables are insigificant, we are not able to clearly differentiate which variable has a larger slice on the pie chart as compared to using a bar chart.
